# **elif statement**

```python
# the above code can be altered using elif statement

party_list = ['SK','RK','RM','UK','UGK']
person = input("Enter the name: ")

if person in party_list:
  print(f"Welcome to the party {person} !!")
elif person not in party_list and person.isupper():
  print(f"Welcome to the party {person} !!")
else:
  print("oops!! you are not invited!")
  
```

```python
#elif statement
#1. Traffic light

color = input("Enter the color you seen on the traffic light: ")

color = color.lower()
if color == 'green':
  print("Car is allowed to go")
elif color == 'yellow':
  print("Car has to wait")
elif color == 'red':
  print("Car has to stop")
else:
  print("unrecognized signal")
```

# **For Loop**

```python
li1 = [1,10,2,20,3,30,4,40,50]
```

```python
for i in li1:
  print(i)
```

```python
for i in li1:
  print(i**2)
```

```python
# add 2
for i in li1:
  print(i+2)
```

```python
for i in li1:
  if i%2 == 0 :
    print("even",i)
  else:
    print("odd",i)
```

```python
# to get sum
s_ = 0
for i in li1:
  s_ += i
print(s_)
```

# ***While Loop***

```python
x = 10
```

```python
while x>0:
  print(x)
  x = x-1
```

```python
# it goes infinite loop
while x>=0:
  if x %2 == 0:
    print("even",x)
  else:
    print("odd",x)
    x = x-1

```

```python
x = 10
while x > len(li1):
  print(x)
  x = x-1
```

```python
# multiple of 5

y = int(input("enter the number :"))

while y>0:
  if y%5 == 0:
    print("multiple of 5")
    break
  else:
    print("not a multiple of 5")
  y = y- 1
```

```python
x = 6
while x >0 :
  if x == 5:
    print("stop!!")
    break
  else:
    print("don't stop")
  x = x-1
```

```python
x = 6
while x >0 :
  if x == 5:
    #print("stop!!")
    continue
  else:
    print("don't stop")
  x = x-1
```
